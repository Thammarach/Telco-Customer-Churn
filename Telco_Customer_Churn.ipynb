```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Dataset_KU/Telco-Customer-Churn - Telco-Customer-Churn.csv')
df
```

```python
df.info()
```

```python
df.isnull().any()
```

```python
df.isna().sum()
```

```python
df.fillna(0, inplace=True)
```

```python
df.set_index('customerID', inplace=True)
df.head()
```

```python
cat_columns = df.select_dtypes('object').columns
cat_columns
```

```python
telcoData_cat = pd.get_dummies(df[cat_columns],drop_first=True)
telcoData_cat
```

```python
num_columns = list(set(df.columns) - set(cat_columns))
num_columns
```

```python
telcoData_prep = pd.concat([df[num_columns],telcoData_cat],axis=1)
```

```python
telcoData_prep
```

# Train test split

```python
from sklearn.model_selection import train_test_split
telcoData_train, telcoData_test  = train_test_split(telcoData_prep, test_size=0.3)
```

```python
telcoData_train
```

```python
pd.crosstab(telcoData_train['Churn_Yes'],columns='Prob',normalize=True) # ตรวจอัตราส่วนดูว่ามี 0 หรือ 1 ใน train มีเท่าไหร่
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=10)
```

```python
knn.fit(telcoData_train.drop(columns='Churn_Yes'), 
         telcoData_train['Churn_Yes'])
```

```python
knn.score(telcoData_train.drop(columns='Churn_Yes'), 
         telcoData_train['Churn_Yes'])
```

```python
knn_tune = KNeighborsClassifier(4)
knn_tune.fit(telcoData_train.drop(columns='Churn_Yes'), 
         telcoData_train['Churn_Yes'])
knn_tune.score(telcoData_train.drop(columns='Churn_Yes'), 
         telcoData_train['Churn_Yes'])
```

# k-fold cross-validation

```python
from sklearn.model_selection import cross_val_score
```

```python
knn10_cv = KNeighborsClassifier(10)
knn5_cv = KNeighborsClassifier(5)

result_knn10 = cross_val_score(knn10_cv,
                               telcoData_train.drop(columns='Churn_Yes'), 
                               telcoData_train['Churn_Yes'], 
                               cv=5)



result_knn5 = cross_val_score(knn5_cv,
                              telcoData_train.drop(columns='Churn_Yes'), 
                              telcoData_train['Churn_Yes'], 
                              cv=5)
```

```python
result_knn10.mean(), result_knn5.mean()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
std_scaler.fit(telcoData_train.drop(columns='Churn_Yes'))
```

```python
Y = telcoData_train['Churn_Yes']
```

```python
std_scaler.mean_, std_scaler.scale_
```

```python
pd.DataFrame(zip(std_scaler.mean_, std_scaler.scale_), index=X.columns, columns=['mean','std']).T
```

```python
X_std_scaled.mean()
```

```python
X_std_scaled.std(ddof=0)
```
